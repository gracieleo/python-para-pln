## Usando Tokenização - para separação de textos, passando parametros especificos: 

* Sem passar parametros, conta vírgula como uma 'palavra'

```python
texto = "A vitória dos Giants, porém, também ficará para a história. Pela primeira vez, irmãos quarterbacks triunfam no Super Bowl em temporadas consecutivas. No ano passado, Peyton Manning, irmão de Eli, chegou ao título máximo da NFL pelo Indianapolis Colts."
nltk.word_tokenize(texto)
```

* Usando Regex, e passando parametros:

```python
from nltk.tokenize import RegexpTokenizer
texto = "A vitória dos Giants, porém, também ficará para a história. Pela primeira vez, irmãos quarterbacks triunfam no Super Bowl em temporadas consecutivas. No ano passado, Peyton Manning, irmão de Eli, chegou ao título máximo da NFL pelo Indianapolis Colts."
tokenizer = RegexpTokenizer(r'[A-z]\w*')     #expressão regular, descartando números e pontos 
tokens = tokenizer.tokenize(texto)
tokens
```

## Usando Frequencia/Contagem de palavras

```python
texto = "A vitória dos Giants, porém, também ficará para a história. Pela primeira vez, irmãos quarterbacks triunfam no Super Bowl em temporadas consecutivas. No ano passado, Peyton Manning, irmão de Eli, chegou ao título máximo da NFL pelo Indianapolis Colts."
tokens = nltk.word_tokenize(texto)
frequencia = nltk.FreqDist(tokens)
frequencia.most_common(10)      #pode passar ou não parametro - pega as top 10 palavras repetidas no texto
```

## Frequencia diferencia maiusculas de minusculas então melhor tratar tudo como minusculas (Lower)

Melhora a frequencia!

* recomendado primeiro tokenizar, depois passar para lower
* palavras iguais porém maiuscula e minuscula contam como sendo diferentes

```python
from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer(r'[A-z]\w*')    #retorna somente palavras 
tokens = tokenizer.tokenize(corpus)

novaLista = []
for token in tokens:
    novaLista.append(token.lower())
    
frequencia = nltk.FreqDist(novaLista)
frequencia.most_common()
```

## Stemmer e Lemmatizer

* stemmer - reduzir palavra ao seu radical -> prop: propõe, propuseram


* lemmatizer - reduzir palavra a forma cacnonica (forma infinitiva), levando em conta classe gramatical -> propor: propõe, propuseram

#### RSLP - Removedor de Sufixos da Lingua Portuguesa    ---->  VARIANTE DE STEMMER 

```python
import nltk
```

```python
stemmer = nltk.RSLPStemmer()            #chamada 

print(stemmer.stem("Amigão"))
print(stemmer.stem("amigos"))
print(stemmer.stem("amiga"))
print(stemmer.stem("propuseram"))        #para verbos não devolve em um mesmo padrão 
print(stemmer.stem("propõe"))
print(stemmer.stem("proposição"))
```

### NLTK - Etiquetadores 

* dizer as classes de palavras 
* em portugues dois corpus que servem como base Floresta, Mac Morpho em ingles existe etiquetador padrão
* então temos que treinar nosso modelo
* resultado ('palavra', 'classe gramatical')


```python
from nltk.corpus import mac_morpho     # base de treinamento
from nltk.tag import UnigramTagger     #pega a classe gramatical mais provavél usando 1 token 

tokens = nltk.word_tokenize(corpus)    # meu texto para tokenizar

sentencas_treino = mac_morpho.tagged_sents()
etiquetador = UnigramTagger(sentencas_treino)

etiquetado = etiquetador.tag(tokens)

print(etiquetado)       
```

* None significa que no corpus de treinamento não existe etiqueta para esse token
* uma alternativa é definir um etiquetador padrão para esses casos de none que deixará tudo não encontrado como substantivo N

```python
from nltk.corpus import mac_morpho     # base de treinamento
from nltk.tag import UnigramTagger     #pega a classe gramatical mais provavél usando 1 token 
from nltk.tag import DefaultTagger 

tokens = nltk.word_tokenize(corpus)    # meu texto para tokenizar

etiq_padrao = DefaultTagger('N')
sentencas_treino = mac_morpho.tagged_sents()
etiquetador = UnigramTagger(sentencas_treino,backoff = etiq_padrao) #caso não encontre um etiquetador padrão use esse

etiquetado = etiquetador.tag(tokens)

print(etiquetado)    
```

#### NLTK Chucking - reconhecimento de entidades nomeadas

```python
from nltk.chunk import RegexpParser

pattern = 'NP: {<NPROP><NPROP> | <N><N>}'  #Retorna par <nome proprio>-<nome proprio> ou <subst.>-<subst.>
analise_gramatical = RegexpParser(pattern)

arvore = analise_gramatical.parse(etiquetado)  #tem que passar o texto já etiquetado
print(arvore)  #arvore pendura os elementos e qdo acha np-np ou n-n gera uma tupla com esse conjunto
```
